```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import plotly.express as px
plt.style.use('default')

from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from kneed import KneeLocator
from sklearn.metrics import silhouette_score
```

```python
# Set the right time format for "Timestamp" column
data['Timestamp'] = pd.to_datetime(data['Timestamp'], format='%d.%m.%Y %H:%M', dayfirst=True)
data
```

```python
# Analyze distribution of the dataset features
features = ['CO2','PM2.5','PM10','Hum','Temp']

fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(17, 2))

for index, var in enumerate(features):
    data[var].plot.hist(ax=axes[index], title=var, bins=12, alpha=0.5)
```

```python
# Preprocessing: scale features distribution - MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data[features])
scaler.data_max_

scaled_data = scaler.transform(data[features])
scaled_data_array = scaler.transform(data[features])

scaled_data = pd.DataFrame(scaled_data, columns=features)

timestamp = data['Timestamp']
scaled_data['Timestamp'] = timestamp
scaled_data = scaled_data[['Timestamp','CO2','PM2.5','PM10','Hum','Temp']]

# Scaled features distribution plot
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(17, 2))

for index, var in enumerate(features):
    scaled_data[var].plot.hist(ax=axes[index], title=var, bins=12, alpha=0.5)
```

```python
# Dataset samples for clustering

scaled_data_CO2vsDust = scaled_data[['CO2','PM2.5','PM10']] # SAMPLE 1 'CO2 vs. Dust'
scaled_data_CO2vsHumTemp = scaled_data[['CO2','Hum','Temp']] # SAMPLE 2 'CO2 vs. HumTemp'
scaled_data_HumvsDust = scaled_data[['Hum','PM2.5','PM10']] # SAMPLE 3 'Hum vs. Dust'

scaled_data_CO2vsDust_array = pd.DataFrame(scaled_data_CO2vsDust).to_numpy() # SAMPLE 1 'CO2 vs. Dust'
scaled_data_CO2vsHumTemp_array = pd.DataFrame(scaled_data_CO2vsHumTemp).to_numpy() # SAMPLE 2 'CO2 vs. HumTemp'
scaled_data_HumvsDust_array = pd.DataFrame(scaled_data_HumvsDust).to_numpy() # SAMPLE 3 'Hum vs. Dust'
```

```python
# SAMPLE 1 'CO2 vs. Dust': Choose the appropriate number of clusters - elbow method
kmeans_kwargs = {
    "init": "random",
    "n_init": 10,
    "max_iter": 300,
    "random_state": 42,
}

sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_data_CO2vsDust_array)
    sse.append(kmeans.inertia_)
    
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()

kl = KneeLocator(range(1, 11), sse, curve="convex", direction="decreasing")
print('k =')
print(kl.elbow)
```

```python
# SAMPLE 1 'CO2 vs. Dust': Choose the appropriate number of clusters - silhouette coefficient
silhouette_coefficients = []

for k in range(2, 11):
     kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
     kmeans.fit(scaled_data_CO2vsDust_array)
     score = silhouette_score(scaled_data_CO2vsDust_array, kmeans.labels_)
     silhouette_coefficients.append(score)
    
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
# SAMPLE 1 'CO2 vs. Dust': k-means clustering 
kmeans = KMeans(
    init="random",
    n_clusters=4,
    n_init=10,
    max_iter=300,
    random_state=42
)

kmeans.fit(scaled_data_CO2vsDust_array)
```

```python
# SAMPLE 1 'CO2 vs. Dust': cluster assignment plot
fig, ax = plt.subplots(nrows=3, figsize=(20,14))

x = data['Timestamp']
label = kmeans.labels_
colors = ['red','green','blue','purple']

for idx, feature in enumerate(['CO2','PM2.5','PM10']):
    y = data[feature]
    ax[idx].scatter(x, y, c=label, cmap=ListedColormap(colors))
    ax[idx].set_title(feature) 
    
plt.show()
fig.savefig('/Users/rodion/Desktop/Sample1.png')
```
